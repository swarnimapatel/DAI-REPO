```python
import pandas as pd
import statsmodels.api as sm
from statsmodels.stats.diagnostic import linear_rainbow

# Step 1: Ensure target variable 'tip' is numeric, converting non-numeric values to NaN
data['tip'] = pd.to_numeric(data['tip'], errors='coerce')

# Step 2: Drop any rows with missing values across the dataset
data = data.dropna()

# Step 3: Separate features (X) and target (y)
X = data.drop(columns='tip')
y = data['tip']

# Step 4: Convert categorical features in X to dummy variables (one-hot encoding)
X_encoded = pd.get_dummies(X, drop_first=True)

# Step 5: Add a constant column for the intercept in the regression model
X_const = sm.add_constant(X_encoded)

# Step 6: Confirm all data is numeric and contains no missing values
# Print if any non-numeric or missing values are still present
print("Any non-numeric values in predictors:", X_const.applymap(lambda x: isinstance(x, (int, float))).all().all())
print("Any missing values in predictors:", X_const.isnull().sum().sum())
print("Any missing values in target:", y.isnull().sum())

# Step 7: Fit the OLS model with processed data
try:
    model = sm.OLS(y, X_const).fit()
    rainbow_stat, rainbow_p_value = linear_rainbow(model)
    print(f'Rainbow test statistic: {rainbow_stat}, p-value: {rainbow_p_value}')
except ValueError as e:
    print("Error in fitting OLS model:", e)
```

```python
# Load and preprocess data
data['tip'] = pd.to_numeric(data['tip'], errors='coerce')
data = data.dropna()
X = data.drop(columns='tip')
y = data['tip']
X_encoded = pd.get_dummies(X, drop_first=True)
X_const = sm.add_constant(X_encoded)

# Check if data is ready for OLS model
print("All numeric in predictors:", X_const.applymap(lambda x: isinstance(x, (int, float))).all().all())
print("No missing values in predictors:", X_const.isnull().sum().sum() == 0)
print("No missing values in target:", y.isnull().sum() == 0)

# Fit OLS model if data is valid
try:
    model = sm.OLS(y, X_const).fit()
    print("Model successfully created.")
    
    # Perform the Rainbow test
    rainbow_stat, rainbow_p_value = linear_rainbow(model)
    print(f'Rainbow test statistic: {rainbow_stat}, p-value: {rainbow_p_value}')
    
    # Residual Plot
    sns.residplot(x=model.fittedvalues, y=model.resid, lowess=True)
    plt.xlabel('Fitted Values')
    plt.ylabel('Residuals')
    plt.title('Residual Plot')
    plt.show()
    
except ValueError as e:
    print("Error in creating model:", e)
```

```python
#preparing the data for modelling 
# One-hot encode categorical variables
X_encoded = pd.get_dummies(X, drop_first=True)

# Split data into training and testing sets
X_train, X_test, y_train, y_test = train_test_split(X_encoded, y, test_size=0.2, random_state=42)
```

```python
#defining models to evaluate 
models = {
    'Linear Regression': LinearRegression(),
    'Ridge Regression': Ridge(alpha=1.0),
    'Lasso Regression': Lasso(alpha=0.1),
    'Decision Tree': DecisionTreeRegressor(),
    'Random Forest': RandomForestRegressor(n_estimators=100, random_state=42),
    'SVR': SVR(),
    'KNN': KNeighborsRegressor(n_neighbors=5)
}
```

```python
#Fit, predict, and evaluate each model
model_results = {}
for model_name, model in models.items():
    model.fit(X_train, y_train)
    y_pred = model.predict(X_test)
    rmse = mean_squared_error(y_test, y_pred, squared=False)
    r2 = r2_score(y_test, y_pred)
    model_results[model_name] = {'RMSE': rmse, 'R^2 Score': r2}
```

```python
# Display results for each model
print("\nModel Evaluation Results:")
for model_name, metrics in model_results.items():
    print(f"{model_name} - RMSE: {metrics['RMSE']:.3f}, R^2 Score: {metrics['R^2 Score']:.3f}")
```
