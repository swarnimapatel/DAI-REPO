```python
import matplotlib.pyplot as plt
import seaborn as sns

# Step 1: Define a function to identify and remove outliers using the IQR method
def remove_outliers(df, column):
    # Calculate Q1 (25th percentile) and Q3 (75th percentile)
    Q1 = df[column].quantile(0.25)
    Q3 = df[column].quantile(0.75)
    IQR = Q3 - Q1  # Interquartile Range

    # Define lower and upper bounds
    lower_bound = Q1 - 1.5 * IQR
    upper_bound = Q3 + 1.5 * IQR

    # Remove rows with outliers
    filtered_df = df[(df[column] >= lower_bound) & (df[column] <= upper_bound)]
    return filtered_df

# Step 2: Visualize the outliers before removal using boxplots
plt.figure(figsize=(15, 5))
for i, column in enumerate(['Total Price', 'Unit Price', 'Quantity'], 1):
    plt.subplot(1, 3, i)
    sns.boxplot(data=data, y=column)
    plt.title(f'Boxplot of {column}')
plt.tight_layout()
plt.show()

# Step 3: Remove outliers from the relevant columns
columns_with_outliers = ['Total Price', 'Unit Price', 'Quantity']
cleaned_data =data.copy()
for col in columns_with_outliers:
    cleaned_data = remove_outliers(cleaned_data, col)

# Step 4: Display the cleaned data info and visualize the boxplots after outlier removal
cleaned_data.info()

plt.figure(figsize=(15, 5))
for i, column in enumerate(['Total Price', 'Unit Price', 'Quantity'], 1):
    plt.subplot(1, 3, i)
    sns.boxplot(data=cleaned_data, y=column)
    plt.title(f'Boxplot of {column} (After Outlier Removal)')
plt.tight_layout()
plt.show()
```

```python
from sklearn.preprocessing import LabelEncoder
import seaborn as sns
import matplotlib.pyplot as plt

# Check data types of all columns and identify non-numeric ones
non_numeric_cols =data.select_dtypes(include=['object']).columns

# Convert all non-numeric columns using Label Encoding
encoded_data = data.copy()
label_encoder = LabelEncoder()

# Apply Label Encoding to each non-numeric column
for col in non_numeric_cols:
    encoded_data[col] = label_encoder.fit_transform(encoded_data[col])

# Calculate pairwise correlation after encoding
correlation = encoded_data.corr()

# Display the correlation matrix
plt.figure(figsize=(12, 10))
sns.heatmap(correlation, annot=True, cmap='coolwarm', fmt='.2f')
plt.title('Correlation Heatmap Including All Variables')
plt.show()
```

```python
import pandas as pd
import scipy.stats as stats

# Group data by 'Product Type' and get the list of 'Total Price' for each group
product_groups = [group['Total Price'].values for name, group in data.groupby('Product Type')]

# Perform one-way ANOVA test
f_stat, p_value = stats.f_oneway(*product_groups)

# Display the results
print(f"F-statistic: {f_stat:.2f}")
print(f"P-value: {p_value:.4f}")

# Interpretation
if p_value < 0.05:
    print("The average Total Price differs significantly across Product Types.")
else:
    print("No significant difference in the average Total Price across Product Types.")
```

```python
# Group data by 'Payment Method' and get the list of 'Total Price' for each group
payment_groups = [group['Total Price'].values for name, group in data.groupby('Payment Method')]

# Perform one-way ANOVA test
f_stat, p_value = stats.f_oneway(*payment_groups)

# Display the results
print(f"F-statistic: {f_stat:.2f}")
print(f"P-value: {p_value:.4f}")

# Interpretation
if p_value < 0.05:
    print("The average Total Price differs significantly across Payment Methods.")
else:
    print("No significant difference in the average Total Price across Payment Methods.")
```

```python
from scipy import stats

# Separate the data into two groups based on Gender
male_prices = data[data['Gender'] == 'Male']['Total Price']
female_prices = data[data['Gender'] == 'Female']['Total Price']

# Perform the independent T-test
t_stat, p_value = stats.ttest_ind(male_prices, female_prices)

# Display the results
print(f"T-statistic: {t_stat:.2f}")
print(f"P-value: {p_value:.4f}")

# Interpretation
if p_value < 0.05:
    print("The average Total Price differs significantly between Male and Female customers.")
else:
    print("No significant difference in the average Total Price between Male and Female customers.")
```
